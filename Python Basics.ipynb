**Python Basics**

**1. Variables**

```python
# A variable stores a piece of data and gives it a name

result  = 124

print(type(result)) # type() function is used for debugging purposes, which returns the type of an object.
# result contained an integer because we gave it an integer

result_1 = result+1.203
print(type(result_1)) # result_1 is a floating point number, or a number containing digits after the decimal point

is_it_Saturday = True
is_it_Monday = False
# These both are 'booleans' or true/false values

my_name = 'Lily'
# This is a string datatype, the name coming from a string of characters

# p.s., we can print all of these with a print command. For example,
print(my_name)
print(f'Is it Saturday? {is_it_Saturday}')



```

**2. More Complicated Data Types**

```python
# What if we want to store many integers or objects? We need a list! 
shopping_cart = ['apple', 'milk', 'beef', 'chicken']
price = [20, 20, 100, 50]

# We can also make an empty list and add objects to it.
color_list = []

color_list.append("Red")
color_list.append("Green")
color_list.append('Blue')

print(color_list)
print(len(color_list)) # len() method returns the length of the list

# We can add any object to a list
color_list.append(12)
print(color_list)

# We can add an object into the specific position in a list
color_list.insert(0, "Orange") # Notice here how the first element of the list is index 0, not 1!
print(color_list)

# We can access a specific emelement of a list too:
print(color_list[0])

print(type(color_list[4]))
print(type(color_list[0]))

# We can change the elemnts in the list
color_list[0] = 'White'
print(color_list)


# In addition to lists, there are tuples
# Tuples behave very similar to lists except that you can't change them after you make them

# An empty tuple isn't very useful:
empty_tuple = ()

# Nor is a tuple with just one element
one_tuple = ('one')
print(one_tuple)


# But tuples with many elements are useful:
students_info = ('Emma', 'HK Island', 2010, 'Februray', 14)

# You can access tuples just like lists
print(f'{students_info[0]} was born in {students_info[2]}.')

# You cannot modifiy existing tuples, but you can make new tuples that extend the information
```

**3. Using Variables**

```python
number_1 = 593.13
number_2 = 2.139

# Addition, subraction, multiplication, division are as you expect

print(number_1 + number_2)
print(number_1 - number_2)
print(number_1 * number_2)
print(number_1 / number_2)

# Modulus
print(9 % 2)

# Powers
print(3 **2)
```

**4. Importing in Python: Math and Plotting**

```python
# 'math' package in Python provides a set of mathematical functions like the ones you have on a calculator. 

import math

print(math.log(number_1))
print(math.exp(number_2))
print(math.pow(3,2))  # The same as 3**2
print(math.fabs(-number_1))
print(math.sqrt(number_1))

# Some special mathematical constants
print(math.pi)  # The mathematical constant pi = 3.1415926...
print(math.e)  # the mathematical constant Euler's number e = 2.718281...
print(math.inf) # a floating-point positive infinity; equivalent to the output of float('inf')

# We can expand the math to entire lists
list_1 = [1, 2, 3, 4, 5, 6]
new_list = [x/2 for x in list_1]
print(new_list)

new_list = [x +2 for x in list_1]
print(new_list)
```

```python
# We can plot easily in Python using 'matplotlib' package

import matplotlib.pyplot as plt

x_vals = [-2, -1, 0, 1, 2]
y_vals = [-6, -3, 0, 3, 6]
plt.plot(x_vals, y_vals)
```

**5. Loops in Python**

```python
# Repeat code until a conditional statement ends the loop

# Let's try print the following list:
number_list = [2, 4, 103 , 76, 3]

# While loops are the basic type
print('While loop:')
i = 0
while i<len(number_list):
  print(number_list[i])
  i  += 1 # the same as i = i+1

# for loop
print('For Loop:')
for i in range(0, len(number_list)): # range(x,y) returns a sequence of numbers from 0 to len(number_list)-1
  print(number_list[i])

print('neater way:')
for x in number_list:
  print(x)
```
